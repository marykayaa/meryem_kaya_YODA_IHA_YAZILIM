```python
# 1. Rastgele Ã–rnekler Ãœzerinde Tahmin (Inference) Yapma

from ultralytics import YOLO
import glob
import random

# EÄŸitim klasÃ¶r adÄ±
TRAIN_RUN_NAME = 'drone_detection_v13'

# En iyi eÄŸitilmiÅŸ modeli yÃ¼klÃ¼yoruz
BEST_MODEL_PATH = f'/content/runs/detect/{TRAIN_RUN_NAME}/weights/best.pt'
trained_model = YOLO(BEST_MODEL_PATH)

# Test setinden rastgele 3 gÃ¶rÃ¼ntÃ¼ seÃ§me
TEST_IMAGES_PATH = '/content/test/images/*.jpg'
all_test_images = glob.glob(TEST_IMAGES_PATH)
num_samples = min(3, len(all_test_images))
random_samples = random.sample(all_test_images, num_samples)

print(f"\nSeÃ§ilen rastgele {len(random_samples)} gÃ¶rÃ¼ntÃ¼ Ã¼zerinde tahmin yapÄ±lÄ±yor...")

# Tahmin yapma (save=True, sonuÃ§larÄ± final_inference klasÃ¶rÃ¼ne kaydeder)
inference_results = trained_model(random_samples, save=True, name=f'final_inference_{TRAIN_RUN_NAME}')

print(f"\nTespit sonuÃ§larÄ± baÅŸarÄ±yla kaydedildi! runs/detect/final_inference_{TRAIN_RUN_NAME} klasÃ¶rÃ¼nÃ¼ kontrol edin.")
```
